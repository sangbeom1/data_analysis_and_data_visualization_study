```python
import pandas as pd
sunspot = pd.read_csv('Yearly Sunspot Number.csv')
```

```python
sunspot['date'] = pd.to_datetime(sunspot['year'].\
                    astype(str).map(lambda x: x+'0101 00:00:00'))

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot('date','sunspot number', '-', data=sunspot)
ax.set_xlabel('date')
ax.set_ylabel('Sunspot Number')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(50,5))
ax = fig.add_subplot(111)
ax.plot('date','sunspot number', '-', data=sunspot, linewidth=7)
ax.set_xlabel('date', size=40)
ax.set_ylabel('Sunspot Number', size=30)
ax.tick_params(axis='both', which='major', labelsize=30)
plt.show()
```

```python
import matplotlib.pyplot as plt
year_range = [range(1700, 1800), range(1800, 1900), range(1900, 2000)]

for i, r in enumerate(year_range):

    fig = plt.figure(figsize=(30,10))
    ax = fig.add_subplot('41' + str(i))
    ax.plot('date','sunspot number', '-', data=sunspot[lambda x: x['year'].isin( r )], linewidth=7)
    ax.set_xlabel('date', size=40)
    ax.set_ylabel('Sunspot Number', size=30)
    ax.tick_params(axis='both', which='major', labelsize=30)

    plt.show()
```
